```python
import sys
import os
from gdocs_service import GoogleDocsService
from utils.markdown_parser import parse_markdown_line
from utils.doc_builder import GoogleDocBuilder
```

```python
def read_meeting_notes():
    """Read meeting notes from markdown file."""
    try:
        with open('meeting_notes.md', 'r') as file:
            return file.read()
    except FileNotFoundError:
        print("‚ùå meeting_notes.md file not found")
        return None
```

```python
def convert_markdown_to_gdoc():
    """Main function to convert markdown to Google Doc."""
    try:
        # Read markdown content
        markdown_content = read_meeting_notes()
        if not markdown_content:
            return
        
        # Parse markdown
        print("üìù Parsing markdown content...")
        lines = markdown_content.split('\n')
        parsed_lines = [parse_markdown_line(line) for line in lines]
        
        # Build requests
        print("üî® Building formatting requests...")
        builder = GoogleDocBuilder()
        requests = builder.build_requests(parsed_lines)
        
        # Initialize Google Docs service
        service = GoogleDocsService()
        
        # Create document
        print("üìÑ Creating new Google Doc...")
        doc_id = service.create_document("Product Team Sync - Converted")
        
        # Apply formatting
        if requests:
            print(f"üé® Applying {len(requests)} formatting operations...")
            service.format_document(doc_id, requests)
            
        # Share document (Optional but recommended for Service Accounts)
        # Set this to your email address to access the doc
        share_email = os.getenv('SHARE_WITH_EMAIL') 
        if share_email:
            print(f"üì§ Sharing document with {share_email}...")
            service.share_document(doc_id, share_email)
        else:
            print("‚ö†Ô∏è Note: Document created by Service Account. Set SHARE_WITH_EMAIL in .env to access it.")
            
        print("‚úÖ Conversion process completed successfully!")
        print(f"üîó Open your document: https://docs.google.com/document/d/{doc_id}")
        return doc_id
        
    except Exception as e:
        print(f"‚ùå Conversion failed: {e}")
        import traceback
        traceback.print_exc()
        return None
```

```python
if __name__ == "__main__":
    doc_id = convert_markdown_to_gdoc()
```
