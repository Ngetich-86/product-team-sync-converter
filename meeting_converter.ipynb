```python
import sys
import os
from src.gdocs_client import GoogleDocsClient
from src.markdown_parser import parse_markdown_line
from src.doc_builder import GoogleDocBuilder
```

```python
def read_meeting_notes():
    """Read meeting notes from markdown file."""
    try:
        with open('meeting_notes.md', 'r') as file:
            return file.read()
    except FileNotFoundError:
        print("‚ùå meeting_notes.md file not found")
        return None
```

```python
def convert_markdown_to_gdoc():
    """Main function to convert markdown to Google Doc."""
    try:
        # Read markdown content
        markdown_content = read_meeting_notes()
        if not markdown_content:
            return
        
        # Parse markdown
        print("üìù Parsing markdown content...")
        lines = markdown_content.split('\n')
        parsed_lines = [parse_markdown_line(line) for line in lines]
        
        # Build requests
        print("üî® Building formatting requests...")
        builder = GoogleDocBuilder()
        requests = builder.build_requests(parsed_lines)
        
        # Initialize Google Docs client
        # Note: This will trigger authentication flow in Colab
        client = GoogleDocsClient()
        
        # Create document
        print("üìÑ Creating new Google Doc...")
        doc_id = client.create_document("Product Team Sync - Converted")
        
        # Apply formatting
        if requests:
            print(f"üé® Applying {len(requests)} formatting operations...")
            client.format_document(doc_id, requests)
            
        print("‚úÖ Conversion process completed successfully!")
        print(f"üîó Open your document: https://docs.google.com/document/d/{doc_id}")
        return doc_id
        
    except Exception as e:
        print(f"‚ùå Conversion failed: {e}")
        import traceback
        traceback.print_exc()
        return None
```

```python
if __name__ == "__main__":
    doc_id = convert_markdown_to_gdoc()
```
